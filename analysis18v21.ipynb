Reset the index to the team for legibility.

```python
nwsl_years = (
    nwsl_full
        .set_index('Team')
        .join(
            nwsl_2018.set_index('Team'),
            lsuffix='21',
            rsuffix='18'
        )
        .reset_index()
)
```

Display shots on goal 2021 vs 2018 next to each other for each team.

```python
nwsl_years = nwsl_years.set_index('Team')
nwsl_years['SOG_per_GP_21'] = nwsl_years['SOG21'] / nwsl_years['GP21']
nwsl_years['SOG_per_GP_18'] = nwsl_years['SOG18'] / nwsl_years['GP18']
nwsl_years[['SOG_per_GP_21', 'SOG_per_GP_18']].head()
```

Goal conversion rate for each team.

```python
nwsl_years['ConvRate_21'] = nwsl_years['G21'] / nwsl_years['SOG21']
nwsl_years['ConvRate_18'] = nwsl_years['G18'] / nwsl_years['SOG18']
nwsl_years[['ConvRate_21', 'ConvRate_18']].head()
```

Display assists per goal for each team.

```python
nwsl_years['Ast_per_G_21'] = nwsl_years['A21'] / nwsl_years['G21']
nwsl_years['Ast_per_G_18'] = nwsl_years['A18'] / nwsl_years['G18']
nwsl_years[['Ast_per_G_18', 'Ast_per_G_21']].head(10)
```

Fouls committed per game.

```python
nwsl_years['FC_per_GP_21'] = nwsl_years['FC21'] / nwsl_years['GP21']
nwsl_years['FC_per_GP_18'] = nwsl_years['FC18'] / nwsl_years['GP18']
nwsl_years[['FC_per_GP_18', 'FC_per_GP_21']].head(10)
```

Number of cards per game for each team.

```python
nwsl_years['Cards_per_GP_21'] = (nwsl_years['YC21'] + nwsl_years['RC21']) / nwsl_years['GP21']
nwsl_years['Cards_per_GP_18'] = (nwsl_years['YC18'] + nwsl_years['RC18']) / nwsl_years['GP18']
nwsl_years[['Cards_per_GP_18', 'Cards_per_GP_21']].head(10)
```

Year over year change for shots on goal and goal conversion rate.

```python
nwsl_years['SOG_change'] = nwsl_years['SOG_per_GP_21'] - nwsl_years['SOG_per_GP_18']
nwsl_years['Conv_change'] = nwsl_years['ConvRate_21'] - nwsl_years['ConvRate_18']
nwsl_years.sort_values('SOG_change', ascending=False)
```

Create a slope chart for the conversion rate from 2018 to 2021. This will show each team's efficiency trend across the two seasons.
Goal conversion rate measures the efficiency of a team's finishing by calculating the percantage of shots on goal that result in an actual goal.
GCR = Goals Scored / Shots on Goal 

```python
nwsl_years = nwsl_years.reset_index()
plt.figure(figsize=(10,6))
for i, row in nwsl_years.iterrows():
    plt.plot([2018, 2021], [row['ConvRate_18'], row['ConvRate_21']], marker='o', label=row['Team'])
plt.xticks([2018, 2021])
plt.ylabel('Goal Conversion Rate')
plt.title('NWSL Teams: Goal Conversion Rate 2018 â†’ 2021')
plt.grid(True, alpha=0.3)
plt.legend(bbox_to_anchor=(1.05,1), loc='upper left')
plt.tight_layout()
plt.show()
```

Create a bar chart to display the number of shots on goal per game for each team in 2021.

```python
plt.figure(figsize=(10,6))
nwsl_years_sorted = nwsl_years.sort_values('SOG_per_GP_21', ascending=False)

plt.bar(nwsl_years_sorted['Team'], nwsl_years_sorted['SOG_per_GP_21'], color='steelblue')
plt.ylabel('Shots on Goal per Game')
plt.title('NWSL Teams: Shots on Goal per Game (2021)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
